## Exploratory Data Analysis

### Import the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the data

```python
df = pd.read_csv(r"C:\Users\pjhop\OneDrive\Documents\Programming & Coding\Python\Projects\Datasets\supermarket_sales.csv")
df.head()
```

### Visuals

```python
sns.set_style("dark")

# Create a figure and two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: Bar chart for City
sns.countplot(x="City", data=df, palette="Blues", ax=axes[0])

# Customize titles and labels for the first subplot
axes[0].set_xlabel("Branch")
axes[0].set_ylabel("Count")
axes[0].set_title("Which Branch is the Most Popular?")
axes[0].grid(linestyle="--", linewidth=.5, color="w")

# Second subplot: Bar chart for Gender
sns.countplot(x="Gender", data=df, palette="Blues", ax=axes[1])

# Customize titles and labels for the second subplot
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Count")
axes[1].set_title("Are there more male or female customers visiting the stores?")
axes[1].grid(linestyle="--", linewidth=.5, color="w")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
sns.set_style("dark")

# Bar Charts
sns.countplot(x="Product line", data=df, palette="Blues", order=df['Product line'].value_counts().index)

# Change titles and labels
plt.xlabel("Product Line")
plt.ylabel("Count")
plt.title("Which Product Line is the Most Popular?")

plt.xticks(rotation=60)

# Grid
plt.grid(linestyle="--", linewidth=.5, color="w")

plt.show()
```

We can see that all the branches have a similar number of customers, with Yangon being slightly most popular than the other two locations. We can see that the most frequent purchases are from the fashion accessories, whilst health and beauty is the least popular, however this could be down to factors such as the cost of the purchase. Therefore in the next step we will investigate whether a particular product line leads to more expensive purchases.

```python
sns.boxenplot(x="gross income", y="Product line", palette="crest", data=df)

# Title and labels
plt.title("Which Product Line leads to the most Gross Income?")
plt.xlabel("Gross Income")
plt.ylabel("Category")

# Display
plt.show()
```

```python
# Add histogram
sns.histplot(x="cogs", data=df, bins=50, alpha=0.5)

# Labels and titles
plt.title("Distribution of purchase prices")
plt.ylabel("Count")
plt.xlabel("Purchase Price")
plt.xlim(0, 1000)

# Add vertical lines
plt.axvline(df['cogs'].mean(), linestyle="--", color="black", label=f"Mean: {df['cogs'].mean():,.2f}")
plt.axvline(df['cogs'].median(), linestyle="--", color="darkgrey", label=f"Median: {df['cogs'].median():,.2f}")

# Add grid
plt.grid(linestyle="--", linewidth=.5, color="white")
plt.legend() # Legend

# Display
plt.show()
```
