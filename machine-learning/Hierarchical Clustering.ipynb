# Hierarchical clustering

## Overview

Hierarchical clustering is a type of unsupervised machine learning algorithm that is used to cluster data based on their similarity or distance. It works differently to K-means clustering in that it does not require a pre-defined number of clusters, 
instead each datapoint is treated as a cluster and then the two most similar are merged until there are no more clusters to merge.

In this project, we will look at how the sklearn iris dataset using the hierarchical clustering algorithm. Our project will take the following format: import of required libraries, import and preparation of the dataset, model fit and evaluation of the model.

## Import of the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram
```

## Import and preparation of the dataset

In this section, we import and scale our data, otherwise if a feature was not scaled and relatively larger compared to the other features, it may lead to poor performance by our model. This is due to how sensitive to the algorithm is to the scale of the features and by scaling, we ensure all features contribute equally to the clustering and there is no bias towards the larger features.

```python
iris = load_iris()
iris_data = iris.data
iris_target = iris.target
```

```python
# Scale the dataset
scaler = StandardScaler()
data_scaled = scaler.fit_transform(iris_data)
```

## Fitting the hierarchical clustering algorithm

```python
# Calculate linkage matrix using Ward's method
WL = linkage(iris_data, method='ward')
CC = linkage(iris_data, method="complete", metric="euclidean")
AC = linkage(iris_data, method="average", metric="euclidean")
SC = linkage(iris_data, method="single", metric="euclidean")
```

The linkage function from scipy is used for hierarchical clustering on the iris dataset, and different linkage methods can be specified. The four methods considered in this case are Ward, Complete, Average, and Single. Ward linkage produces compact and balanced clusters of relatively equal size. Complete linkage can lead to long chains of points that are not very compact. Average linkage tends to produce more balanced and compact clusters than complete linkage. Single linkage can be sensitive to outliers and noise, resulting in long chains of points that may not belong to the same cluster.

`WL, CC, SC, SC` contains all the indices of the clusters that have been merged and the distance between them using the four different methods, which we will visualize in the **dendrograms** below. This will highlight the differences between the methods.

```python
# Plot dendrogram
plt.figure(figsize=(12, 8))
plt.title('Hierarchical Clustering Dendrogram - Ward Linkage')
plt.xlabel('Data Points')
plt.ylabel('Distance')
dendrogram(WL)
plt.show()
```

```python
# Plot dendrogram
plt.figure(figsize=(12, 8))
plt.title('Hierarchical Clustering Dendrogram - Complete Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
dendrogram(CC)
plt.show()
```

```python
# Plot dendrogram
plt.figure(figsize=(12, 8))
plt.title('Hierarchical Clustering Dendrogram - Average Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
dendrogram(AC)
plt.show()
```

```python
# Plot dendrogram
plt.figure(figsize=(12, 8))
plt.title('Hierarchical Clustering Dendrogram - Single Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
dendrogram(SC)
plt.show()
```

Dendrograms are a powerful advantage of hierarchical clustering algorithms, as they allow us to visualize the clusters and understand their similarity levels. By analyzing the dendrogram, we can determine the optimal number of clusters and decide how to group the data in a meaningful way. The height of the dendrogram between two clusters can also be used to determine the dissimilarity level, indicating whether two clusters should be considered as separate clusters. In summary, dendrograms provide valuable insights into the clustering structure of the data and assist in making informed decisions about the number and composition of clusters.
