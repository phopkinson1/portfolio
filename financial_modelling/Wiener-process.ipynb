# Wiener-process

The Wiener-process is a stochastic process that is used to model Brownian motion, which is typically used in finance to predict the movements of stock prices. 

**Formula for Geometric random walk**

$$dS = \mu\cdot S\cdot dt + \sigma\cdot S\cdot dW $$

This equation can be split into two components:
* The first term in the equation ($\mu\cdot S\cdot dt$) is the deterministic component - which represents the expected growth of stock over the given time period. Where $\mu$ is the expected return of the stock, S is the stock price and $dt$ is the time increment. The component can be defined as the expected return of stock over the given time period.
* The second term in the equation ($\sigma\cdot S\cdot dW$) is the stochastic component - which represents the fluctuations or noise of a stock price over a given time period. Here $\sigma$ is the standard deviation of the stock returns, and $dW$ is the Brownian motion or Wiener process. The component can be defined as the randomness or fluctuations in the stock returns.

## Import the required libraries

For this project, we will import the following libraries:
* `numpy` for mathematical operations.
* `numpy.random` for the random generation of values from the normal distribution.
* `matplotlib.pyplot` for the graphing of the wiener-process over time. 

```python
import numpy.random as npr
import numpy as np
import matplotlib.pyplot as plt
```

## Define functions for wiener-process and graph plotting

```python
def wiener_process(dt=0.1, x=0, N=1000):
    # Generate an array of zeros for W
    W = np.zeros(N+1)
    
    # Generates an array of timesteps
    t = np.linspace(x, N, N+1)
    
    # For the wiener process it generates a normal distribution
    W[1:N+1] = np.cumsum(npr.normal(x, np.sqrt(dt), N))
    
    # Returns time and wiener_process values
    return t, W
```

The `wiener-process` function generates a wiener-process, it does this by taking the following steps:
1. The function first takes the following three arguments:
    * `dt` which represents the change in time
    * `x` which represents the mean value of the process
    * `N` which represents the total number of timesteps
2. Inside the function, we define a `W` parameter, which is this case represents our wiener process and the `W = np.zeros(N+1)` will generate a array of zeros for the length of the total number of timesteps include zero. 
3. Next, we define a `t` parameter, which represents the total number of timesteps, with `0` as the starting value, `N` as the end value and `N+1` being the total number of samples to be generated.
4. The next line of code generates the wiener process for all values other than the starting value.
    * `npr.normal` draws random samples from a normal distribution, which has a mean of `x`, standard deviation of `np.sqrt(dt)` and size of `N`.
    * `np.cumsum` then adds together all the values in the array up till and including each N value. 
5. Finally, the code returns both the time and wiener process values. 

```python
def plot_process(t, W):
    # Plot command
    plt.plot(t, W)
    
    # Title and axis labels
    plt.xlabel('Time')
    plt.ylabel('Wiener-process W(t)')
    plt.title('Wiener-process')
    
    # Display
    plt.show()
```

The `plot_process` function works by plotting the timesteps and wiener-process values on a graph, providing labels and then displaying the graph.

## Functions in practice

```python
time, data = wiener_process()
plot_process(time, data)
```
